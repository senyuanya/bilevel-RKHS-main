```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
import os

import sys 
sys.path.append("..") 
import add_mypathlevy
```

```python
def load_results_linear(dx_coarse, dx, dt, N, supp_u,t_range, kernel_type, scale_value, method_our):
    datafolder = os.path.join(add_mypathlevy.data_folder, f"FDM{method_our}_n05x_dx_coarse{dx_coarse}_dx{dx}_dt{dt}_N{N}_supp_u{supp_u[-1]}_t_range{t_range[-1]}_kernel_type{kernel_type}")
    Data_set = np.load(os.path.join(datafolder, f"bikernel_data_scale{scale_value}.npz"),allow_pickle=True)
    Loss_set = np.load(os.path.join(datafolder, f"bikernel_loss_scale{scale_value}.npz"),allow_pickle=True)
    G_set = np.load(os.path.join(datafolder, f"bikernel_Gbar_scale{scale_value}.npz"),allow_pickle=True)
    UF_set = np.load(os.path.join(datafolder, f"bikernel_uxfx_scale{scale_value}.npz"),allow_pickle=True)
    Lnorm_set = np.load(os.path.join(datafolder, f"bikernel_L2l2{scale_value}.npz"),allow_pickle=True)
    return Data_set, Loss_set, G_set,UF_set, Lnorm_set
```

```python
def load_results_sin(dx_coarse, dx, dt, N, supp_u,t_range, kernel_type, scale_value, method_our):
    datafolder = os.path.join(add_mypathlevy.data_folder, f"FDM{method_our}_sinx_dx_coarse{dx_coarse}_dx{dx}_dt{dt}_N{N}_supp_u{supp_u[-1]}_t_range{t_range[-1]}_kernel_type{kernel_type}")
    Data_set = np.load(os.path.join(datafolder, f"bikernel_data_scale{scale_value}.npz"),allow_pickle=True)
    Loss_set = np.load(os.path.join(datafolder, f"bikernel_loss_scale{scale_value}.npz"),allow_pickle=True)
    G_set = np.load(os.path.join(datafolder, f"bikernel_Gbar_scale{scale_value}.npz"),allow_pickle=True)
    UF_set = np.load(os.path.join(datafolder, f"bikernel_uxfx_scale{scale_value}.npz"),allow_pickle=True)
    Lnorm_set = np.load(os.path.join(datafolder, f"bikernel_L2l2{scale_value}.npz"),allow_pickle=True)
    return Data_set, Loss_set, G_set,UF_set, Lnorm_set
```

```python
dx = 0.005# space mesh size for generating data
N = 30 # number of data pairs (u_i, f_i);
u_Type = 'Compoundpdf'  # types: 'Bspline', 'Fourier','Compoundpdf',Laplacepdf; Fourier works for mfOpt, which requires derivatives of u.
x_range = [-5, 5] # space range
dt = 0.000025 # time mesh size in CFL condition
t_range = [0, 1] # time range
example_type = 'nonlocal'  # {'LinearIntOpt', 'nonlocal', 'mfOpt'}
kernel_type = 'Compoundlevy'  # Gaussian, sinkx, FracLap,Compoundlevy,Laplacejump
scale_value = 1
method_our = 'BiSNGD'
supp_u = [0,2]
```

```python
dx_coarse1 = 0.01
dataset1L, lossset1L, Gset1L, UFset1L, Lnorm_set1L = load_results_linear(dx_coarse1,dx, dt, N, supp_u,t_range, kernel_type, scale_value, method_our)
 
iteration_history1L = lossset1L["iteration_history"]
loss_history1L = lossset1L["loss_history"]
error_history1L = lossset1L["error_history"]

dataset1S, lossset1S, Gset1S, UFset1S, Lnorm_set1S = load_results_sin(dx_coarse1,dx, dt, N, supp_u,t_range, kernel_type, scale_value, method_our)
 
iteration_history1S = lossset1S["iteration_history"]
loss_history1S = lossset1S["loss_history"]
error_history1S = lossset1S["error_history"]
```

```python
dx_coarse4 = 0.04
dataset4L, lossset4L, Gset4L, UFset4L, Lnorm_set4L = load_results_linear(dx_coarse4,dx, dt, N, supp_u,t_range, kernel_type, scale_value, method_our)
 
iteration_history4L = lossset4L["iteration_history"]
loss_history4L = lossset4L["loss_history"]
error_history4L = lossset4L["error_history"]

dataset4S, lossset4S, Gset4S, UFset4S, Lnorm_set4S = load_results_sin(dx_coarse4,dx, dt, N, supp_u,t_range, kernel_type, scale_value, method_our)
 
iteration_history4S = lossset4S["iteration_history"]
loss_history4S = lossset4S["loss_history"]
error_history4S = lossset4S["error_history"]
```

```python
log10_loss1L = [np.log10(loss) for loss in loss_history1L]
log10_error1L = [np.log10(error) for error in error_history1L]
log10_loss1S = [np.log10(loss) for loss in loss_history1S]
log10_error1S = [np.log10(error) for error in error_history1S]

log10_loss2L = [np.log10(loss) for loss in loss_history2L]
log10_error2L = [np.log10(error) for error in error_history2L]
log10_loss2S = [np.log10(loss) for loss in loss_history2S]
log10_error2S = [np.log10(error) for error in error_history2S]

log10_loss3L = [np.log10(loss) for loss in loss_history3L]
log10_error3L = [np.log10(error) for error in error_history3L]
log10_loss3S = [np.log10(loss) for loss in loss_history3S]
log10_error3S = [np.log10(error) for error in error_history3S]

log10_loss4L = [np.log10(loss) for loss in loss_history4L]
log10_error4L = [np.log10(error) for error in error_history4L]
log10_loss4S = [np.log10(loss) for loss in loss_history4S]
log10_error4S = [np.log10(error) for error in error_history4S]
```

```python
dxlist1 = [dx_coarse1, dx_coarse2, dx_coarse3, dx_coarse4]

error_logS = [log10_error1S[-1], log10_error2S[-1], log10_error3S[-1], log10_error4S[-1]]
loss_logS = [log10_loss1S[-1], log10_loss2S[-1], log10_loss3S[-1], log10_loss4S[-1]]
error_logL = [log10_error1L[-1], log10_error2L[-1], log10_error3L[-1], log10_error4L[-1]]
loss_logL = [log10_loss1L[-1], log10_loss2L[-1], log10_loss3L[-1], log10_loss4L[-1]]

dxlist1_log = [np.log10(dx_coarse1), np.log10(dx_coarse2), np.log10(dx_coarse3), np.log10(dx_coarse4)]
```

```python
slope_err_logL, intercept_err_logL = np.polyfit(dxlist1_log, error_logL, 1)
slope_loss_logL, intercept_loss_logL = np.polyfit(dxlist1_log, loss_logL, 1)

slope_err_logS, intercept_err_logS = np.polyfit(dxlist1_log, error_logS, 1)
slope_loss_logS, intercept_loss_logS = np.polyfit(dxlist1_log, loss_logS, 1)
```

```python
print("slope_err_logL",slope_err_logL)
print("slope_err_logS",slope_err_logS)
```

```python
mid_idx = slice(1, 3)         #  dxlist1[1], dxlist1[2]
dx_mid = dxlist1_log[mid_idx]

#  ndarray
dx_list = np.array(dxlist1_log, dtype=float)
dx_mid   = np.array(dx_mid,   dtype=float)

# 
y_fit_err_mid_logL  = slope_err_logL  * dx_mid + intercept_err_logL
y_fit_err_mid_logS = slope_err_logS * dx_mid + intercept_err_logS

# 
plt.rcParams.update({
    "figure.figsize": (12, 10),
    "xtick.labelsize": 28,
    "ytick.labelsize": 28,
    "legend.fontsize": 21,
})

#  y 
fig, ax1 = plt.subplots(figsize=(12, 10))

# 
colors      = ['#1f4e79', '#6B8E23']
linestyles  = ['-.', '--']
markers     = ['o', 's']
linewidth_v = 6

#  log10(ErrorL)
ax1.plot(dxlist1_log, error_logL,
         color=colors[0], linestyle=linestyles[0],
         linewidth=linewidth_v, alpha=0.85,
         marker=markers[0], markersize=10, markeredgecolor='black',
         label=r'$b(x) = -0.5x$')

#  log10(ErrorS)
ax1.plot(dxlist1_log, error_logS,
         color=colors[1], linestyle=linestyles[1],
         linewidth=linewidth_v, alpha=0.85,
         marker=markers[1], markersize=10, markeredgecolor='black',
         label=r'$b(x) = \sin x$')

# log10 
ax1.plot(dx_mid-0.05, y_fit_err_mid_logL+0.1 ,
         color='#be0f2d', linestyle='-', linewidth=4) #  label=rf'Fit$_{{err}}$: $slope={slope_err_log:.1e}$'
ax1.plot(dx_mid-0.05, y_fit_err_mid_logS-0.03,
         color='#be0f2d', linestyle='-', linewidth=4) #label=rf'Fit$_{{loss}}$: $slope={slope_loss_log:.1e}$'

err_shiftL = 0.25
err_shiftS = -0.05
x_shift = 0.1
x_text_errL = dx_mid[0] - x_shift +0.1
y_text_errL = y_fit_err_mid_logL[0] + err_shiftL

x_text_errS = dx_mid[0] - x_shift +0.1
y_text_errS = y_fit_err_mid_logS[0] + err_shiftS

ax1.text(x_text_errL, y_text_errL,
         rf'slope:{slope_err_logL.round(2)}', color=colors[0], fontsize=20, fontweight='bold')
ax1.text(x_text_errS, y_text_errS,
         rf'slope:{slope_err_logS.round(2)}', color=colors[1], fontsize=20, fontweight='bold')


# Labels
ax1.set_xlabel(r'$\log_{10}(\Delta x)$', fontsize=30, fontweight='bold') #= \{\log_{10}(0.01), \log_{10}(0.02), \log_{10}(0.025), \log_{10}(0.04)\}
ax1.set_ylabel(r'$\log_{10}(L_2(\rho)\mathrm{Error})$', fontsize=30, fontweight='bold')

#  min(log) 
ax1.set_ylim(bottom= min(np.min(error_logL),np.min(error_logS )) - 0.1)

y1min, y1max = ax1.get_ylim()
ticks1 = np.arange(-2.5,-0.5,0.5)

ax1.set_yticks(ticks1)


# 
ax1.tick_params(axis='x', labelsize=28)
ax1.tick_params(axis='y', labelsize=28)


# 
lines1, labels1 = ax1.get_legend_handles_labels()

ax1.legend(lines1 , labels1 , frameon=False, loc='upper right', fontsize=21)

# 
#ax1.grid(True, linestyle='--', alpha=0.6, color='gray')

plt.tight_layout()

figure_path = os.path.join(add_mypathlevy.figure_folder, f"{kernel_type}AllLossErrordx{dx_coarse1}N{N}.pdf")
plt.savefig(figure_path, bbox_inches='tight', dpi=500)
plt.show()
```
